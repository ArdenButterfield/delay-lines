```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import scipy.fftpack as fp
```

```python
im = np.random.random((10,10))
```

```python
def makeFiltered(im, minimum, maximum):
    F1 = fp.fft2((im).astype(float))
    F2 = fp.fftshift(F1)

    (w, h) = im.shape
    half_w, half_h = int(w/2), int(h/2)
    
    # high pass filter
    F2[half_w - minimum : half_w + minimum + 1, half_h - minimum : half_h + minimum + 1] = 0 # select all but the first 50x50 (low) frequencies
    F2[:half_w - maximum, :] = 0
    F2[half_w + maximum:, :] = 0
    F2[:, :half_h - maximum] = 0
    F2[:, half_h + maximum:] = 0
    im1 = fp.ifft2(fp.ifftshift(F2)).real
    return im1
```

```python
with open('FilteredSquares.cpp', 'w') as f:
    f.write('#include "FilteredSquares.h"\n#include <array>\n#include <unordered_map>\n\n')
    f.write(f'const std::unordered_map<int, std::array<unsigned char, {im.shape[0]} * {im.shape[1]}>> filteredSquares = {{\n')
    for minimum in range(0, 25, 1):
        for maximum in range(minimum + 1, 26, 1):
            print(minimum, maximum)
            f.write(f'\t{{{minimum * 1000 + maximum}, {{')
            filtered = makeFiltered(im, minimum, maximum)
            f.write(','.join((filtered.flatten() * 256).astype(np.uint8).astype(str)))
            f.write('}},\n')
    f.write('};\n')
    
```
